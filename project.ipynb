#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
sublist['normalized_price'] = sublist.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())
locations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
plt.figure(figsize=(10, 6))

for location in locations:
    plt.hist(sublist[sublist['level5'] == location]['normalized_price'], bins=20, label=location)

plt.title('Histograms of Normalized Prices by Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
from ipyleaflet import Map, basemaps, CircleMarker

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
colors = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "purple"
}

nuevo_ds = ds[ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]
nuevo_ds['latitude'] = nuevo_ds['latitude'].astype("string")
nuevo_ds['longitude'] = nuevo_ds['longitude'].astype("string")
nuevo_ds['latitude'] = nuevo_ds['latitude'].replace(',', '.', regex=True)
nuevo_ds['longitude'] = nuevo_ds['longitude'].replace(',', '.', regex=True)
coordinates = nuevo_ds[["latitude", "longitude"]].values
population_sb = nuevo_ds["level5"].values

for (lat,long),pop in zip(coordinates,population_sb): 
        if lat != "0" and long != "0":
            circle = CircleMarker(
            location=(lat, long),
            radius=5,
            color=colors[pop],
            fill=True,
            fill_opacity=0.7

            )
    


            map.add_layer(circle)
        else: 
            continue
        
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps, CircleMarker, LayerGroup


# Center the map around Madrid (approx. latitude 40.4, longitude -3.7)
map = Map(center=(40.4, -3.7), zoom=10, min_zoom=1, max_zoom=20, 
          basemap= basemaps.OpenStreetMap.Mapnik)

# Define a color for each population
colors = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "purple"
}

# Create a LayerGroup to hold all CircleMarkers
marker_group = LayerGroup()

# Plot each population on the map
for index, row in sb_count.iterrows():
    circle = CircleMarker(
        location=(row['latitude'], row['longitude']),
        radius=5,
        color=colors[row['level5']],
        fill=True,
        fill_opacity=0.7
    )
    marker_group.add_layer(circle)

# Add the marker group to the map
map.add_layer(marker_group)

map
```
